```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import os
import pprint
import csv
from api_keys import api_key
from citipy import citipy
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# Starter code
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)
        
len(cities)
```

```python
def Weather_Data(cities, apikey):
    url = "http://api.openweathermap.org/data/2.5/weather"
    payload  = {"appid": api_key,
                "q": cities,
                "units": "imperial"
               }
    r = requests.get(url, params=payload)
    if r.status_code == 200:
        return r.json()
    else:
        return None
```

```python
print("Processing data for:")
Weather_Data(cities, api_key)
data = []
city_num = 1
for city in cities:
    result = Weather_Data(city, api_key)
    if result:
        print ("City Number " + str(city_num) + ": " + (city))
        data.append(result)
        city_num += 1
```

```python
Weather_Frame = pd.DataFrame([{
                "City Name": d["name"],
                "Latitude": d["coord"]["lat"],
                "Maximum Temperature (f)": d["main"]["temp_max"],
                "Humidity (%)": d["main"]["humidity"],
                "Cloudiness(%)": d["clouds"]["all"],
                "Wind Speed (mph)": d["wind"]["speed"]
    } for d in data])
```

```python
Weather_Frame.to_csv("Weather_Frame.csv", index = False)
```

#### Latitude vs. Temperature Plot

```python
x = Weather_Frame["Latitude"]
y = Weather_Frame["Maximum Temperature (f)"]
plt.scatter(x, y, marker = "o", edgecolors = "black")
plt.xlabel("Degrees of Latitude")
plt.ylabel("Maximum Temperatures (f)")
plt.title ("City Latitude v. Maximum Temperatures (f) on 1/5/2020")
plt.grid(axis = "y")
plt.grid(axis = "x")
plt.xlim([-90, 90])
plt.rcParams["figure.figsize"] = (15,15)
plt.savefig("Temp.png")
plt.show()
```

## Trends

1) The more northerly (positive) the latitude, the colder the temperatures. The more southerly (negative) the latitude, the warmer the temperatures. This makes sense given it is winter in the northern hemisphere and summer in the southern hemisphere on 1/5/2020.

2) There are no observable trends related to latitude and humidity, wind speed, or cloud cover.

3) The sampled cities are relatively more humid than not no matter the location in the world
